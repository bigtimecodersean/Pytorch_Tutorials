```python
import torch
```

Temperature Model

```python
t_c = [0.5,  14.0, 15.0, 28.0, 11.0,  8.0,  3.0, -4.0,  6.0, 13.0, 21.0]
t_u = [35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4]
t_c = torch.tensor(t_c)
t_u = torch.tensor(t_u)
```

```python
#define model 
def model(t_u, w1, w2, b):
    return w2*t_u**2+w1*t_u+b
```

```python
#define loss function
def loss_fn(t_p, t_c):
    squared_diffs = (t_p - t_c)**2
    return squared_diffs.mean()
```

```python
w1 = torch.ones(())
w2 = torch.ones(())
b = torch.zeros(())
t_p = model(t_u, w1, w2, b)
t_p
```

```python
#check value of loss
loss = loss_fn(t_p, t_c)
loss
```

```python
delta = 0.1
loss_rate_of_change_w1 = \
(loss_fn(model(t_u, w1 + delta,w2, b), t_c) - loss_fn(model(t_u, w1 - delta,w2, b), t_c)) / (2.0 * delta)
```

```python
print(loss_rate_of_change_w1)
```

```python
learning_rate = 1e-2
w1 = w1 - learning_rate * loss_rate_of_change_w1
```

```python
#same with w2

loss_rate_of_change_w2 = \
(loss_fn(model(t_u, w1,w2 + delta, b), t_c) - loss_fn(model(t_u, w1, w2+delta, b), t_c)) / (2.0 * delta)
w2 = w2 - learning_rate * loss_rate_of_change_w2
```

```python
#same with b
```

```python
loss_rate_of_change_b = \
    (loss_fn(model(t_u, w1, w2, b + delta), t_c) -
     loss_fn(model(t_u, w1, w2, b - delta), t_c)) / (2.0 * delta)
b = b - learning_rate * loss_rate_of_change_b
```

```python
#Defining the gradient function
def dloss_fn(t_p, t_c):
    dsq_diffs = 2 * (t_p - t_c) / t_p.size(0)
    return dsq_diffs

def dmodel_dw1(t_u, w1, w2, b):
  return t_u

def dmodel_dw2(t_u, w1, w2, b):
  return t_u*2*w2


def dmodel_db(t_u, w1, w2, b):
  return 1.0


def grad_fn(t_u, t_c, t_p, w1, w2, b):
  dloss_dtp = dloss_fn(t_p, t_c)
  dloss_dw1 = dloss_dtp * dmodel_dw1(t_u, w1, w2, b)
  dloss_dw2 = dloss_dtp * dmodel_dw2(t_u, w1, w2, b)
  dloss_db = dloss_dtp * dmodel_db(t_u, w1, w2, b)
  return torch.stack([dloss_dw1.sum(), dloss_dw2.sum(), dloss_db.sum()])
```

```python
#define training loop
def training_loop(n_epochs, learning_rate, params, t_u, t_c):
  for epoch in range(1, n_epochs + 1):
    w1, w2, b = params
    t_p = model(t_u, w1, w2, b)
    loss = loss_fn(t_p, t_c)
    grad = grad_fn(t_u, t_c, t_p, w1, w2, b)
    params = params - learning_rate * grad
    print('Epoch %d, Loss %f' % (epoch, float(loss)))
  return params
```

```python
#Execute training loop
training_loop(
    n_epochs = 100,
    learning_rate = 1e-2,
    params = torch.tensor([1.0, 1.0, 0.0]),
    t_u = t_u,
    t_c = t_c)
```

Rewrite using autograd 

```python

def model(t_u, w1, w2, b):
   return w2*t_u**2+w1*t_u+b


def loss_fn(t_p, t_c):
   squared_diffs = (t_p - t_c)**2
   return squared_diffs.mean()

params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)

```

```python
loss = loss_fn(model(t_u, *params), t_c)
loss.backward()
```

```python
params.grad
```

At this point, the grad attribute of params contains the derivatives of the loss with respect to each element of params.

```python
#Full autograd enabled 
def training_loop(n_epochs, learning_rate, params, t_u, t_c):
  for epoch in range(1, n_epochs + 1):
      if params.grad is not None:
          params.grad.zero_()

      t_p = model(t_u, *params)
      loss = loss_fn(t_p, t_c)
      loss.backward()

      with torch.no_grad():
       params -= learning_rate * params.grad

      if epoch % 500 == 0:
                    print('Epoch %d, Loss %f' % (epoch, float(loss)))

  return params

```

```python
training_loop(
  n_epochs = 5000,
  learning_rate = 1e-2,
  params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True),
  t_u = t_un,
  t_c = t_c)
```

```python
#update the training loop
def training_loop(n_epochs, optimizer, params, t_u, t_c):
    for epoch in range(1, n_epochs + 1):
        t_p = model(t_u, *params)
        loss = loss_fn(t_p, t_c)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if epoch % 500 == 0:
            print('Epoch %d, Loss %f' % (epoch, float(loss)))
    return params
```

```python
n_samples = t_u.shape[0]
n_val = int(0.2 * n_samples)
shuffled_indices = torch.randperm(n_samples)
train_indices = shuffled_indices[:-n_val]
val_indices = shuffled_indices[-n_val:]
train_indices, val_indices
```

```python
train_t_u = t_u[train_indices]
train_t_c = t_c[train_indices]
val_t_u = t_u[val_indices]
val_t_c = t_c[val_indices]
train_t_un = 0.1 * train_t_u
val_t_un = 0.1 * val_t_u
```

```python
def training_loop(n_epochs, optimizer, params, train_t_u, val_t_u,
              train_t_c, val_t_c):
  for epoch in range(1, n_epochs + 1):
      train_t_p = model(train_t_u, *params)
      train_loss = loss_fn(train_t_p, train_t_c)
      val_t_p = model(val_t_u, *params)
      val_loss = loss_fn(val_t_p, val_t_c)

      optimizer.zero_grad()
      train_loss.backward()
      optimizer.step()

      if epoch <= 3 or epoch % 500 == 0:
              print(f"Epoch {epoch}, Training loss {train_loss.item():.4f},"
                    f" Validation loss {val_loss.item():.4f}")
  return params

```

```python
params = torch.tensor([1.0, 1.0, 0.0], requires_grad=True)
learning_rate = 1e-5
optimizer = optim.SGD([params], lr=learning_rate)


training_loop(
    n_epochs = 3000,
    optimizer = optimizer,
    params = params,
    train_t_u = train_t_un,
    val_t_u = val_t_un,
    train_t_c = train_t_c,
    val_t_c = val_t_c)
```
